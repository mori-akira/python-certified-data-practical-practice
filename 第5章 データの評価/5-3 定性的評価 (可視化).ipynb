```python
# pandasを使った可視化
import pandas as pd

df = pd.read_parquet("../result/penguins.parquet")
df.plot(
    x="Culmen_Length",
    y="Flipper_Length",
    kind="scatter",
)
```

```python
df.plot(
    y="Body_Mass",
    kind="hist",
)
```

```python
df.plot()
```

```python
# Matplotlibを用いた可視化
import matplotlib.pyplot as plt

_, ax = plt.subplots()
ax.scatter(
    x=df.loc[:, "Culmen_Length"],
    y=df.loc[:, "Flipper_Length"],
)
plt.show()
```

```python
# plotlyを用いた可視化
import plotly.express as px

px.scatter(
    df,
    x="Culmen_Length",
    y="Flipper_Length",
)
```
