```python
# csvモジュール
import csv

with open("../data/人口推計.csv", encoding="utf-8", newline="") as f:
    csv_reader = csv.reader(f)

    for i, row in enumerate(csv_reader):
        print("| ", end="")
        for col in row:
            print(col, end=" | ")
        print()
        if i >= 4:
            break
        print("=====")
```

```python
# 主な事前定義のdialect
def print_dialect_info(dialect_name):
    try:
        dialect = csv.get_dialect(dialect_name)
        print(f"Dialect: {dialect_name}")
        print(f"  delimiter: {repr(dialect.delimiter)}")
        print(f"  quotechar: {repr(dialect.quotechar)}")
        print(f"  doublequote: {dialect.doublequote}")
        print(f"  skipinitialspace: {dialect.skipinitialspace}")
        print(f"  lineterminator: {repr(dialect.lineterminator)}")
        print(f"  quoting: {dialect.quoting}")
    except csv.Error:
        print(f"Error: Dialect '{dialect_name}' not found.")

for dialect_name in ["excel", "unix", "my_dialect"]:
    print_dialect_info(dialect_name)
    print("-" * 40)
```

```python
# カスタム定義
csv.register_dialect(
    "my_dialect",
    delimiter="|",
    quotechar="'",
    doublequote=False,
    skipinitialspace=True,
    lineterminator="\n",
)
print_dialect_info("my_dialect")
```
