```python
import warnings
from scipy.io import wavfile

warnings.filterwarnings("ignore")
filename = "../data/sample.wav"
samplingrate, data = wavfile.read(filename)
print(type(samplingrate))
print(samplingrate)
```

```python
print(type(data))
print(data.dtype)
print(data.shape)
```

```python
data[:16]
```

```python
# 音声ファイルの長さ
length = data.shape[0] / samplingrate
length
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Matplotlibで可視化
t = np.linspace(0.0, length, data.shape[0]) # 等差数列(min, max, count)
plt.plot(t, data[:, 0])
plt.xlabel("time")
plt.ylabel("amplitude")
```

```python
plt.plot(t, data[:, 1], color="orange")
plt.xlabel("time")
plt.ylabel("amplitude")
```
