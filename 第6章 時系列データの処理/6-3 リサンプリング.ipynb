```python
# resample()メソッド
import pandas as pd

ser = pd.Series(
    range(21),
    index=pd.date_range("2025-01-01", periods=21),
)
ser.head()
```

```python
ser.resample("1W")
```

```python
ser.resample("1W").first()
```

```python
ser.resample("1W").sum()
```

```python
ser.resample("1W").ohlc()
```

```python
ser.resample("3D").ohlc()
```

```python
ser.resample("3D", closed="left").ohlc()
```

```python
ser.resample("3D", closed="right").ohlc()
```

```python
pd.Series(
    range(21),
    index=pd.date_range(
        "2025-01-01",
        periods=21,
        freq="1W",
    ),
).resample("1D").ohlc().head(15)
```

```python
pd.Series(
    range(21),
    index=pd.date_range(
        "2025-01-01",
        periods=21,
        freq="1W",
    ),
).resample("1D").ohlc().ffill().head(15)
```
