```python
from IPython.display import display
import pandas as pd

df1 = pd.DataFrame(
    [
        [1, 10, 100],
        [2, 20, 200],
        [3, 30, 300],
    ],
    index=list("abc"),
    columns=list("ABC"),
)
df2 = pd.DataFrame(
    [
        [4, 10, 100],
        [2, 20, 200],
        [3, 30, 400],
    ],
    index=list("abd"),
    columns=list("ABD"),
)
display(df1, df2)
```

```python
# concat()による連結
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis=0, join="outer")
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
pd.concat([df1, df2], keys=["data1", "data2"])
```

```python
pd.concat([df2, df1], join="inner")
```

```python
# concatによる結合
pd.concat([df1, df2], axis=1)
```

```python
pd.concat([df2, df1], axis=1)
```

```python
pd.concat([df2, df2], axis=1)
```

```python
pd.concat([df2, df1], join="inner", axis=1)
```

```python
df3 = pd.DataFrame(
    [
        [1.3, 10.3],
        [2.3, 20.3],
    ],
    index=list("ab"),
    columns=list("DE"),
)
display(df1, df2, df3)
```

```python
# join()による結合
df1.join(df3)
```

```python
df1.join(df2)
```

```python
df1.join(
    df2,
    how="inner",
    lsuffix="_df1",
    rsuffix="_df2",
)
```

```python
df1.join(
    df2,
    how="outer",
    rsuffix="_df2",
)
```

```python
df4 = pd.DataFrame(
    [
        ["a", 1.4, 10.4],
        ["b", 2.4, 20.4],
    ],
    columns=["key", "F", "G"],
)
display(df4, df1)
```

```python
df4.join(df1, on="key")
```

```python
# merge()による結合
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on="A")
```

```python
pd.merge(df1, df2, on="B")
```

```python
pd.merge(df1, df2, on=["A", "B"], how="right")
```

```python
pd.merge(df1, df2, how="outer")
```

```python
pd.merge(df2, df1, on="A", how="outer", suffixes=("-df1", "-df2"), indicator=True)
```

```python
pd.merge(df1, df2, how="cross", indicator=True)
```

```python
pd.merge(df1, df2, on="A", how="cross", indicator=True)
```

```python
# 欠損値を含む同士の結合
import numpy as np

pd.merge(
    pd.DataFrame({"a": [np.nan, 1, 2, np.nan], "b": ["あ", "い", "う", "え"]}),
    pd.DataFrame({"a": [1, np.nan, 1], "b": ["か", "き", "く"]}),
    on="a",
)
```

```python
pd.merge(
    pd.DataFrame({"a": [np.nan, 1, 2, np.nan], "b": ["あ", "い", "う", "え"]}),
    pd.DataFrame({"a": [1, np.nan, 1], "b": ["か", "き", "く"]}),
    on="a",
    validate="one_to_one"
)
```

```python
pd.merge(
    pd.DataFrame({"a": [np.nan, 1, 2, 3], "b": ["あ", "い", "う", "え"]}),
    pd.DataFrame({"a": [1, 1, np.nan, np.nan], "b": ["か", "き", "く", "け"]}),
    on="a",
    validate="one_to_many"
)
```

```python
# 近い値での結合
df1 = pd.DataFrame(
    [
        [90.9, 80.8, 70.7],
        [100.3, 123.4, 150.01],
    ]
)
df2 = pd.DataFrame(
    [
        [90.1, 80.2 ,70.3],
        [100.2, 151.02, 143.2],
    ]
)
pd.merge_asof(df1, df2, on=0)
```

```python
pd.merge_asof(df1, df2, on=1, tolerance=1)
```

```python
pd.merge_asof(df1, df2, left_on=0, right_on=1, tolerance=10.8)
```
